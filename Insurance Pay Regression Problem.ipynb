```python
def sex_change(x):
    d={"male":1,"female":0}
    if x in d.keys():
        return d[x]
    
```

```python
health["sex"]=health["sex"].apply(sex_change)
```

```python
def smoke_change(x):
    d={"yes":1,"no":0}
    if x in d.keys():
        return d[x]
```

```python
health["smoker"]=health["smoker"].apply(smoke_change)
```

```python
health.corr()
```

```python
sns.heatmap(health.corr(),annot=True)
```

```python
X=health[['age', 'sex', 'bmi', 'children', 'smoker']]
```

```python
y=health['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50)
#most of histogram is at 0 which means it is working fine
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
